````python
import requests
import time
from datetime import date,datetime,timezone
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Settings
finn_key = "btv5stv48v6q7nvmtsjg"
base_url = "https://finnhub.io/api/v1/stock/candle"
````

````python
# This function pulls and plots daily stock prices for a company ('symbol') over a given date range and overlays Bollinger Band chart indicators
def linearreg(symbol, start_date, end_date):
    # Format API call with passed in parameters
    params = {'symbol': symbol,
              'resolution': 'D',
              'from': start_date,
              'to': end_date,
              'token': finn_key}
    response = requests.get(base_url,params=params).json()
    
    # Convert API response to dataframe
    df = pd.DataFrame(response)
    
    # Create date column that is human readable
    df['date'] = [datetime.utcfromtimestamp(d) for d in df['t']] # utcfromtimestamp for accurate date
    # Typical price is the average of the high, low and close
    df['tp'] = (df.h + df.l + df.c) / 3.0
    
    df.set_index(pd.DatetimeIndex(df['date']), inplace=True) # needs datetime as index for rolling offset
    
    # 1 day moving average of the typical price
    df['ma1']= df.tp.rolling('1D').mean() # 1 day offset

    # Create a plot using plotly with a plot for the moving average, 
    # a candlestick plot for the stock we're interested in
#     fig = go.Figure(data = [
#             go.Scatter(x=df['date'],y=df['ma1'],line=dict(color='gray',dash='dash'),name='1 Day Moving Average'),
#             go.Candlestick(
#             x=df['date'],
#             open=df['o'],
#             high=df['h'],
#             low=df['l'],
#             close=df['c'],
#             name=symbol
#         )
#     ])
    fig = px.scatter(x=df['date'],y=df['ma1'], trendline="ols")

    fig.update_layout(
    title="{} with 5-year Linear Regression".format(symbol),
    xaxis_title="Date",
    yaxis_title="Price ($)",
#     legend_title="Legend Title",
    )
    # Allow for adjusting the y-axis
    fig.update_yaxes(fixedrange = False)
    return fig
````

````python
stocks = ['AMZN','AAPL','MSFT','TSLA']

end_date = 1603238400 # Oct 21, 2020
# start_date = end_date - 5 years
start_date = end_date-5*365*24*3600-86400 # Oct 21, 2018 need to subtract extra day

for stock in stocks:
    fig = linearreg(stock, start_date, end_date)
    fig.show()
````
