```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
apple_df = pd.read_csv("data/apple.csv")
apple_df.head()
```

```python
apple_df.columns=['Date', 'Close/Last', 'Volume', 'Open', 'High', 'Low']
```

```python
apple_df['High']=apple_df['High'].str.replace('$','')
apple_df['High']=pd.to_numeric(apple_df['High'])
apple_df.dtypes
```

```python
apple_df[['Date', 'High']].set_index('Date').plot()
plt.ylabel('Highest Price')
plt.title('Apple Highest Prices over the Years')
plt.savefig('Highest Prices over the Years from Apple.png')
```
