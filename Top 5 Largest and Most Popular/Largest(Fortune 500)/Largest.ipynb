```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
```

```python
#Read in data
amazon = pd.read_csv('data/AMZN.csv')
apple = pd.read_csv('data/AAPL.csv')
walmart = pd.read_csv('data/WMT.csv')
exxon = pd.read_csv('data/XOM.csv')
cvs = pd.read_csv('data/CVS.csv')
```

```python
#Set Figure Size
fig = plt.figure()
plt.rcParams["figure.figsize"] = [20,12]

#Plot Data points
ax = amazon[['Date', 'Close']].set_index('Date').plot(color='b', label='Amazon')
apple[['Date', 'Close']].set_index('Date').plot(ax=ax, color='r', label='Apple')
walmart[['Date', 'Close']].set_index('Date').plot(ax=ax, color='y', label='Walmart')
exxon[['Date', 'Close']].set_index('Date').plot(ax=ax, color='g', label='Exxon')
cvs[['Date', 'Close']].set_index('Date').plot(ax=ax, color='purple', label='CVS')

#Add legend
plt.legend(['Amazon', "Apple", 'Walmart', "Exxon", 'CVS'], loc='best')

#Format Graph
plt.yticks(np.arange(0, 3800, 200.0))
```
