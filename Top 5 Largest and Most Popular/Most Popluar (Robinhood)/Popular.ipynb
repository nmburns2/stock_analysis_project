```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import glob
```

```python
#Read in data
aphria = pd.read_csv('data/APHA.csv')
alibaba = pd.read_csv('data/BABA.csv')
catalyst = pd.read_csv('data/CPRX.csv')
sony = pd.read_csv('data/SNE.csv')
virgin_galactic = pd.read_csv('data/SPCE.csv')
```

```python
#Set Figure Size
fig = plt.figure()
plt.rcParams["figure.figsize"] = [20,12]

#Plot Data points
ax = aphria[['Date', 'Close']].set_index('Date').plot(color='b', label='Aphria')
alibaba[['Date', 'Close']].set_index('Date').plot(ax=ax, color='r', label='Alibaba')
catalyst[['Date', 'Close']].set_index('Date').plot(ax=ax, color='y', label='Catalyst Pharmaceuticals')
sony[['Date', 'Close']].set_index('Date').plot(ax=ax, color='g', label='Sony')
virgin_galactic[['Date', 'Close']].set_index('Date').plot(ax=ax, color='purple', label='Virgin Galactic')

#Add legend
plt.legend(['Aphria', "Alibaba", 'Catalyst Pharmaceuticals', "Sony", 'Virgin Galactic'], loc='best')

#Format Graph
plt.yticks(np.arange(0, 400, 25))
```
